```python
import pandas as pd
import pickle
from fastapi import FastAPI, HTTPException, File, UploadFile
from pydantic import BaseModel
from model_utils import transform_and_select_features, selected_features
from typing import Union
from fastapi.encoders import jsonable_encoder
import json
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
import numpy as np
from typing import Any
def transform_and_select_features(data):
    # Identify columns with a single null value
    columns_to_impute = data.columns[data.isnull().sum() == 1].tolist()
    print(columns_to_impute)
    # Create a copy of the DataFrame for imputation
    data_imputed = data.copy()

    # Apply imputation only to columns with a single null value
    for column in columns_to_impute:
        if column in data_imputed.columns:
            imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
            data_imputed[column] = imputer.fit_transform(data[[column]])

    return data_imputed

# Example usage
data = pd.DataFrame({
    'col1': [1, 2, np.nan, 4]
})

transformed_data = transform_and_select_features(data)
print("Original Data:")
print(data)

print("\nTransformed Data:")
print(transformed_data)
```
